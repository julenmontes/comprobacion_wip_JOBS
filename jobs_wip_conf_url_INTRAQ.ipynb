```python
wip_confs = wip.sort_values(by='version_simple',ascending=False)[['implantation','conf','ConfigURL','Extralibs','NombreJob','version','version_simple']]

wip_confs['Extralibs'] = np.where(
    wip_confs['conf'].duplicated(keep=False),
    wip_confs.merge(wip_confs.dropna(subset=['conf','Extralibs']).rename(columns={'Extralibs':'Extralibs1'}).drop_duplicates('conf'),how='left', on='conf').Extralibs1,
    wip_confs['Extralibs']
    )

wip_confs = wip_confs.drop_duplicates(subset=['conf']).reset_index(drop=True)

wip_confs['NombreJobFinal'] = np.where(
    wip_confs['NombreJob'].duplicated(keep=False),
    wip_confs['NombreJob'] + '0' + wip_confs.groupby('NombreJob').cumcount().add(1).astype(str).str[-2:],
    wip_confs['NombreJob'] + '01'
    )
wip_confs = wip_confs.drop(columns=['NombreJob'])
wip_confs['UA'] = wip_confs['conf'].str[:4].str.lower()
wip_confs['len_NombreJobFinal'] = wip_confs.NombreJobFinal.str.len()

if not UUAAs:
    UUAAs = list(wip_confs.conf.str[:4].str.lower().sort_values().unique())
wip_confs.query('UA == @UUAAs', inplace=True) #nos quedamos con nuestras UUAAs

wip_confs
```
