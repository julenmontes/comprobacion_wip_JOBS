## Tiempo de integración

```python
data = pd.read_csv('./tm_integracion.csv')
data.tail(2)
```

```python
limite = 60*24 #un día
data_median = data.groupby('repository_url')['integration_time (min)'].median().sort_values(ascending=False).reset_index()
repos_limite = data_median[data_median['integration_time (min)']>limite]
if len(repos_limite)==0:
    print('TODOS LOS REPOS ESTAN OK')
else:
    print('¡REVISAR LOS SIGUIENTES REPOS!:')
for i in repos_limite.values:
    print('-mediana:',i[1],'\t',i[0])
```
